Case: A university wants a system where teachers can calculate: 1. Total marks for 3 subjects, or 2. 4 subjects (for labs), or 3. Any number of subjects (optional course included) Task: Create a class Grade with an overloaded total_marks () method using: Default parameters or *args. Calculate percentage for the student based on number of subjects provided.

```python
class Grade:
    def total_marks(self, *marks):
        total = sum(marks)
        num_subjects = len(marks)
        percentage = (total / (num_subjects * 100)) * 100 
        return total, percentage
grade = Grade()
total_3, percent_3 = grade.total_marks(85, 90, 78)
print(f"Total Marks (3 subjects): {total_3}, Percentage: {percent_3:.2f}%") 
total_4, percent_4 = grade.total_marks(85, 90, 78, 88)
print(f"Total Marks (4 subjects): {total_4}, Percentage: {percent_4:.2f}%")
total_n, percent_n = grade.total_marks(85, 90, 78, 88, 92, 76,87,43,90)
print(f"Total Marks (n subjects): {total_n}, Percentage: {percent_n:.2f}%")
```
