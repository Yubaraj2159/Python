```python
# Dictionary with student names as keys and marks as values
students = {
    "Manisha": 85,
    "Prabin": 92,
    "Amit": 78,
    "Gaurav": 95,
    "Anish": 88
}

# Calculate average marks
average_marks = sum(students.values()) / len(students)
print(f"Average marks: {average_marks:.2f}")  //Infront f is the formating f   //Back f is for digits after decimal.

# Find the top scorer
top_scorer = max(students, key=students.get)
top_score = students[top_scorer]
print(f"Top scorer: {top_scorer} with {top_score} marks")
```
