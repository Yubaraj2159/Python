Create a base class Vehicle with attributes make and model. Derive a class Car from Vehicle with additional attributes like mileage and capacity. Create objects of both classes and display their details.


```python

class Vehicle:
    def __init__(self, make: str, model: str):
        self.make = make
        self.model = model

    def details(self) -> str:
        return f"Make: {self.make}, Model: {self.model}"

    def __str__(self) -> str:
        return self.details()


class Car(Vehicle):
    def __init__(self, make: str, model: str, mileage: float, capacity: int):
        super().__init__(make, model)
        self.mileage = mileage
        self.capacity = capacity

    def details(self) -> str:
        return f"{super().details()}, Mileage: {self.mileage}, Capacity: {self.capacity}"

    def __str__(self) -> str:
        return self.details()


# Create objects and display details
v = Vehicle("Toyota", "Corolla")
c = Car("Honda", "Civic", mileage=18.5, capacity=5)

print("Vehicle:", v)
print("Car:    ", c)
```
