Create a class MaxFinder that identifies the largest number in a list.

```python

class MaxFinder:
    def __init__(self, numbers):
        self.numbers = numbers
    
    def find_max(self):
        return max(self.numbers)

a = [5, 9, 10, 20, 25, 8, 13]
finder = MaxFinder(a)
print(f"The largest number is: {finder.find_max()}")
```
