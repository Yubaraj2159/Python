Write a Python class called Student with attributes like name, roll number, and marks. Create methods to display student details and calculate the average marks.


```python
class Student:
    def __init__(self, name, roll_number, marks=None):
        """
        Initialize a Student.

        :param name: student's name (str)
        :param roll_number: student's roll number (int or str)
        :param marks: iterable of numeric marks (list/tuple)
        """
        self.name = name
        self.roll_number = roll_number
        self.marks = list(marks) if marks is not None else []

    def display_details(self):
        """Print student details and average marks."""
        avg = self.average_marks()
        print(f"Name: {self.name}")
        print(f"Roll Number: {self.roll_number}")
        print(f"Marks: {self.marks}")
        print(f"Average: {avg:.2f}" if self.marks else "Average: N/A")

    def average_marks(self):
        """Return the average of marks (0.0 if no marks)."""
        if not self.marks:
            return 0.0
        return sum(self.marks) / len(self.marks)


# Example usage (remove or adapt when using in notebook)
if __name__ == "__main__":
    s = Student("Alice", 101, [85, 92, 78])
    s.display_details()
    print("Average (returned):", s.average_marks())
```
