```python
# dictionary of students and their attendance percentages
students_attendance = {
    "Alice": 82,
    "Bob": 72,
    "Charlie": 66,
    "Diana": 90,
    "Eve": 74,
}

threshold = 75

# print students with attendance below threshold
low_attendance = {name: pct for name, pct in students_attendance.items() if pct < threshold}
if low_attendance:
    for name, pct in low_attendance.items():
        print(f"{name}: {pct}%")
else:
    print("No students with attendance below 75%") 
```
