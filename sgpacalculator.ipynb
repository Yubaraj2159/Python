```python
grades = ["A", "A-", "A", "A", "A"]
credits = [3, 3, 3, 3, 3]

grade_points = {
    "A": 4.0,
    "A-": 3.7,
    "B+": 3.3,
    "B": 3.0,
    "B-": 2.7,
    "C+": 2.3,
    "C": 2.0,
    "C-": 1.7,
    "D+": 1.3,
    "D": 1.0,
    "F": 0
}

def calculate_sgpa(grades, credits):
    total_points = 0
    total_credits = 0

    for i in range(len(grades)):
        grade = grades[i].upper()
        if grade in grade_points:
            total_points += grade_points[grade] * credits[i]
            total_credits += credits[i]
        else:
            print(f"Invalid grade '{grade}' found. Skipping this subject.")

    if total_credits == 0:
        return 0

    sgpa = total_points / total_credits
    return sgpa

# Calculate SGPA
sgpa = calculate_sgpa(grades, credits)
print(f"\nYour SGPA is: {sgpa:.2f}")
```
